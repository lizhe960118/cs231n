```python
import time
import numpy as np
import matplotlib.pyplot as plt
from cs231n.classifiers.fc_net import *
from cs231n.data_utils import get_CIFAR10_data
from cs231n.gradient_check import eval_numerical_gradient, eval_numerical_gradient_array
from cs231n.solver import Solver

%matplotlib inline
plt.rcParams['figure.figsize'] = (8.0, 12.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

%load_ext autoreload
%autoreload 2

def rel_error(x, y):
  """ returns relative error """
  return np.max(np.abs(x - y) / (np.maximum(1e-8, np.abs(x) + np.abs(y))))
```

```python
data = get_CIFAR10_data()
for k,v in data.items():
    print('%s' % k, v.shape)
```

## Batch Normalization:forward

```python
# initial a two layer network
N, D1, D2, D3 = 200, 50, 60, 3
X = np.random.randn(N, D1)
W1 = np.random.randn(D1, D2)
W2 = np.random.randn(D2, D3)
a = np.maximum(0, X.dot(W1)).dot(W2)

print('Before batch normalization')
print('  means:', a.mean(axis=0))
print('  stds:', a.std(axis=0))

print('\nAfter batch normalization(gamma=1, beta=0)')
a_norm, _ =  batchnorm_forward(a, np.ones(D3), np.zeros(D3),{'mode':'train'})
print('  means:', a_norm.mean(axis=0))
print('  stds:', a_norm.std(axis=0))

gamma = np.array([1.0, 2.0, 3.0])
beta = np.array([11.0, 12.0, 13.0])
print('\nAfter batch normalization(nontrivial gamma. beta)')
a_norm, _ =  batchnorm_forward(a, gamma, beta,{'mode':'train'})
print('  means:', a_norm.mean(axis=0))
print('  stds:', a_norm.std(axis=0))
```

## Check the means and variances of activations in test forward pass

```python
N, D1, D2, D3 = 200, 50, 60, 3
W1 = np.random.randn(D1, D2)
W2 = np.random.randn(D2, D3)

gamma = np.ones(D3)
beta = np.zeros(D3)

bn_param = {'mode': 'train'}
for t in range(50):
    X = np.random.randn(N, D1)
    a = np.maximum(0, X.dot(W1)).dot(W2)
    a_norm, _  = batchnorm_forward(a, gamma, beta, bn_param)
#     print('  means:', a_norm.mean(axis=0))
#     print('  stds:', a_norm.std(axis=0))

bn_param['mode'] = 'test'
X = np.random.randn(N, D1)
a = np.maximum(0, X.dot(W1)).dot(W2)
a_norm, _ =  batchnorm_forward(a, gamma, beta, bn_param)
    
print('Before batch normalization, mean and variances in test time')
print('  means:', a_norm.mean(axis=0))
print('  stds:', a_norm.std(axis=0))
```

## Batch Normalization：backward

```python
N, D = 4, 5
x = 5 * np.random.randn(N, D) + 12
gamma = np.random.randn(D)
beta = np.random.randn(D)
dout = np.random.randn(N, D)

bn_param = {'mode':'train'}

fx = lambda _:batchnorm_forward(x, gamma, beta, bn_param)[0]
fg = lambda _:batchnorm_forward(x, gamma, beta, bn_param)[0]
fb = lambda _:batchnorm_forward(x, gamma, beta, bn_param)[0]

dx_num = eval_numerical_gradient_array(fx, x, dout)
dg_num = eval_numerical_gradient_array(fg, gamma, dout)
db_num = eval_numerical_gradient_array(fb, beta, dout)

_, cache = batchnorm_forward(x, gamma, beta, bn_param)
dx, dgamma, dbeta = batchnorm_backward(dout, cache)

print('the relative error in %s is %e' % ('dx', rel_error(dx, dx_num)))
print('the relative error in %s is %e' % ('dgamma', rel_error(dgamma, dg_num)))
print('the relative error in %s is %e' % ('dbeta', rel_error(dbeta, db_num)))

# _, cache = batchnorm_forward(x, gamma, beta, bn_param)
dx, dgamma, dbeta = batchnorm_backward_alt(dout, cache)

print('the relative error in %s is %e' % ('dx', rel_error(dx, dx_num)))
print('the relative error in %s is %e' % ('dgamma', rel_error(dgamma, dg_num)))
print('the relative error in %s is %e' % ('dbeta', rel_error(dbeta, db_num)))
```

## Compare time in batchnorm_backward and batchnorm_backward_alt

```python
import time
N, D = 100, 500
x = 5 * np.random.randn(N, D) + 12
gamma = np.random.randn(D)
beta = np.random.randn(D)
dout = np.random.randn(D)

bn_param = {'mode':'train'}
out, cache = batchnorm_forward(x, gamma, beta, bn_param)

print("using the batchnorm_backward")
t_start = time.time()
dx1, dgamma1, dbeta1 = batchnorm_backward(dout, cache)
t_end = time.time()
t1 = t_end - t_start
print('time used is %f ms' % t1)

print("using the batchnorm_backward_alt")
t_start = time.time()
dx2, dgamma2, dbeta2 = batchnorm_backward_alt(dout, cache)
t_end = time.time()
t2 = t_end - t_start
print('time used is %f ms' % t2)

print('the relative error in %s is %e' % ('dx', rel_error(dx1, dx2)))
print('the relative error in %s is %e' % ('dgamma', rel_error(dgamma1, dgamma2)))
print('the relative error in %s is %e' % ('dbeta', rel_error(dbeta1, dbeta2)))
print('time speedup is %f' % ((t1 - t2)/ t1))
```

## Use BN in FC_net

```python
N, D, H1, H2, C = 2, 15, 20, 30 ,10
X = np.random.randn(N,D)
y = np.random.randint(C, size=(N, ))

for reg in [0, 3.14]:
    print('Running check with reg : %0.2f' % reg)
    model = FullyConnectedNet([H1, H2], 
                              input_size=D,
                              output_size=C, 
                              reg=reg, 
                              weight_scale=5e-2,
                              dtype=np.float64, 
                              use_batchnorm=True)
    loss, grads = model.loss(X, y)
    print('Initial loss: %0.6f' % loss)
    
    # gradient check
    for name in sorted(grads):
        f = lambda _:model.loss(X, y)[0]
        grad_num = eval_numerical_gradient(f, model.params[name], verbose=False)
        print('%s relative error: %.2e' % (name, rel_error(grad_num, grads[name])))
```

## Batchnorm for deep networks

```python
num_train = 1000

small_data = {
    'X_train':data['X_train'][:num_train],
    'y_train':data['y_train'][:num_train],
    'X_val':data['X_val'],
    'y_val':data['y_val']
}

weight_scale = 2e-2
learning_rate = 1e-3

hidden_dims = [100, 100, 100, 100, 100]

model = FullyConnectedNet(hidden_dims, 
                          weight_scale=weight_scale, 
                          dtype=np.float64)
bn_model = FullyConnectedNet(hidden_dims, 
                             weight_scale=weight_scale, 
                             dtype=np.float64, 
                             use_batchnorm=True)

solver = Solver(model, small_data,
                update_rule = 'Adam',
                optim_param={'learning_rate':learning_rate},
                num_epochs=10,
                batch_size=50,
                p_num=200,
                verbose=True)
solver.train()

bn_solver = Solver(bn_model, small_data,
                   update_rule = 'Adam',
                   optim_param={'learning_rate':learning_rate},
                   num_epochs=10,
                   batch_size=50,
                   p_num=200,
                   verbose=True)
bn_solver.train()
```

```python
plt.subplot(3, 1, 1)
plt.title('Training loss')
plt.xlabel('Iteration')

plt.subplot(3, 1, 2)
plt.title('Training accuracy')
plt.xlabel('Epoch')

plt.subplot(3, 1, 3)
plt.title('Validation accuracy')
plt.xlabel('Epoch')

plt.subplot(3, 1, 1)
plt.plot(solver.loss_history, 'o', label='without BN')
plt.plot(bn_solver.loss_history, 'o', label='with BN')

plt.subplot(3, 1, 2)
plt.plot(solver.train_acc_history, '-o', label='without BN')
plt.plot(bn_solver.train_acc_history, '-o', label='with BN')

plt.subplot(3, 1, 3)
plt.plot(solver.val_acc_history, '-o', label='without BN')
plt.plot(bn_solver.val_acc_history, '-o', label='with BN')

for i in [1, 2, 3]:
    plt.subplot(3, 1, i)
    plt.legend(loc='upper center', ncol=4)

plt.gcf().set_size_inches(15, 15)
plt.show()
```

## iteraction of BN and weight initialization

```python
weight_scales = np.logspace(-4, 0, num=20)
weight_scales
```

```python
num_train = 1000

small_data = {
    'X_train':data['X_train'][:num_train],
    'y_train':data['y_train'][:num_train],
    'X_val':data['X_val'],
    'y_val':data['y_val']
}

weight_scales = np.logspace(-4, 0, num=20)
learning_rate = 1e-3

hidden_dims = [50, 50, 50, 50, 50, 50, 50]

bn_solvers = {}
solvers = {}

for i, weight_scale in enumerate(weight_scales):
    print('Running the experiment : %d / %d, The weight_scale is %e' % (i+1, len(weight_scales), weight_scale))
    model = FullyConnectedNet(hidden_dims, 
                              weight_scale=weight_scale, 
                              dtype=np.float64)
    bn_model = FullyConnectedNet(hidden_dims, 
                                 weight_scale=weight_scale, 
                                 dtype=np.float64, 
                                 use_batchnorm=True)

    solver = Solver(model, small_data,
                    update_rule = 'Adam',
                    optim_param={'learning_rate':learning_rate},
                    num_epochs=10,
                    batch_size=50,
                    p_num=200,
                    verbose=False)
    solver.train()
    solvers[weight_scale] = solver

    bn_solver = Solver(bn_model, small_data,
                       update_rule = 'Adam',
                       optim_param={'learning_rate':learning_rate},
                       num_epochs=10,
                       batch_size=50,
                       p_num=200,
                       verbose=False)
    bn_solver.train()
    bn_solvers[weight_scale] = bn_solver
```

```python
final_train_loss, bn_final_train_loss = [], []
best_train_accs, bn_best_train_accs = [], []
best_val_accs, bn_best_val_accs = [], []

for weight_scale in weight_scales:
    final_train_loss.append(np.mean(solvers[weight_scale].loss_history[-100:]))
    best_train_accs.append(max(solvers[weight_scale].train_acc_history))
    best_val_accs.append(max(solvers[weight_scale].val_acc_history))

    bn_final_train_loss.append(np.mean(bn_solvers[weight_scale].loss_history[-100:]))
    bn_best_train_accs.append(max(bn_solvers[weight_scale].train_acc_history))
    bn_best_val_accs.append(max(bn_solvers[weight_scale].val_acc_history))

plt.subplot(3, 1, 1)
plt.title('Train Loss')
plt.xlabel('weight_scale')

plt.subplot(3, 1, 2)
plt.title('Best train accuracy')
plt.xlabel('weight_scale')

plt.subplot(3, 1, 3)
plt.title('Best validation accuracy')
plt.xlabel('weight_scale')

plt.subplot(3, 1, 1)
plt.plot(final_train_loss, '-o', label='without BN')
plt.plot(bn_final_train_loss, '-o', label='with BN')

plt.subplot(3, 1, 2)
plt.plot(best_train_accs, '-o', label='without BN')
plt.plot(bn_best_train_accs, '-o', label='with BN')

plt.subplot(3, 1, 3)
plt.plot(best_val_accs, '-o', label='without BN')
plt.plot(bn_best_val_accs, '-o', label='with BN')

for i in [1, 2, 3]:
    plt.subplot(3, 1, i)
    plt.legend(loc='upper center', ncol=4)

plt.gcf().set_size_inches(15, 15)
plt.show()
```

可以看出BN有助于减少权重取值的影响，至于权重初始化可随意取值并不太现实。
